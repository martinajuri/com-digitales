```python
import random
import numpy as np
import matplotlib.pyplot as plt

numIter = np.arange(100,20000,200)
Pa =np.zeros(len(numIter))
Pb = np.zeros(len(numIter))
Pab = np.zeros(len(numIter))
Pbgivena = np.zeros(len(numIter))
k = 0
for veces in numIter:
    Na, Nb, Nab = 0, 0, 0
    dado1 = np.zeros(veces)
    dado2 = np.zeros(veces)
    
    for N in range(veces):
        dado1[N] = random.randint(1,6)
        dado2[N] = random.randint(1,6)
        if (dado1[N] + dado2[N]) % 2 == 0:
            Na+=1 
            if (dado1[N] + dado2[N]) % 4 == 0:
                Nab+=1
        if (dado1[N] + dado2[N]) % 4 == 0:
            Nb+=1

    Pa[k] = Na / N
    Pb[k] = Nb / N
    Pab[k] = Nab / N
    Pbgivena[k]= Pab[k] / Pa[k] if Pa[k] > 0 else 0

    k+=1

plt.plot(numIter,Pab)
plt.xlabel('Nmero de iteraciones')
plt.ylabel('Na/N')
plt.title('Clculo de la probabilidad por frecuencia relativa Pab')
plt.grid()
plt.show()
```

```python
plt.plot(numIter,Pbgivena)
plt.xlabel('Nmero de iteraciones')
plt.ylabel('Na/N')
plt.title('Clculo de la probabilidad por frecuencia relativa Pb dado a')
plt.grid()
plt.show()
```

```python
print("La ultima probabilidad calculada es de: ",Pab[-1])
print("La probablidad terica de : ", 1/4)
```

```python
print("La ultima probabilidad calculada es de: ",Pbgivena[-1])
print("La probablidad terica de : ", 1/2)
```

```python
print("Los eventos A y B son independientes si P(AB) = P(A)*P(B)")
print("P(AB) es", Pab[-1])
print("P(A) es", Pa[-1])
print("P(B) es", Pb[-1])
print("P(A)*P(B) es", Pa[-1]*Pb[-1])
print("Por lo tanto, los eventos A y B son dependientes")
```
