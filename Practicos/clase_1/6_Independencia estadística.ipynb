```python
import random
import numpy as np
import matplotlib.pyplot as plt

numIter = np.arange(100,20000,200)
Pa =np.zeros(len(numIter))
k = 0
for veces in numIter:
    Na = 0
    dado1 = np.zeros(veces)
    dado2 = np.zeros(veces)
    for N in range(veces):
        dado1[N] = random.randint(1,6)
        dado2[N] = random.randint(1,6)
        if dado1[N] == dado2[N]==5:
            Na+=1 
    Pa[k]= Na/N
    k+=1

plt.plot(numIter,Pa)
plt.xlabel('Nmero de iteraciones')
plt.ylabel('Na/N')
plt.title('Clculo de la probabilidad por frecuencia relativa')
plt.grid()
plt.show()
```

```python
print("La ultima probabilidad calculada es de: ",Pa[-1])
print("La probablidad terica de : ", 1/36)
```
