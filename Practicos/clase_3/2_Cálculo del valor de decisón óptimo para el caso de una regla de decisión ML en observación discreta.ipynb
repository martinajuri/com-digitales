```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_samples = 100000

H = np.random.randint(0, 2, num_samples)  # Hipótesis transmitida
N = np.random.randint(0, 2, num_samples)  # Ruido aditivo

Y = H + N

P_Y_H_0 = [np.mean(Y[H == 0] == y) for y in range(3)]
P_Y_H_1 = [np.mean(Y[H == 1] == y) for y in range(3)]

decision_rule = np.argmax([P_Y_H_0, P_Y_H_1], axis=0)

def calcular_error(umbral):
    H_decidido = (Y >= umbral).astype(int)
    return np.mean(H_decidido != H)

Pe_0 = calcular_error(0)
Pe_1 = calcular_error(1)
Pe_2 = calcular_error(2)
```

```python
print(f"Probabilidades condicionales P(Y|H=0): {P_Y_H_0}")
print(f"Probabilidades condicionales P(Y|H=1): {P_Y_H_1}")
print(f"Probabilidad de error para umbral 0: {Pe_0:.4f}")
print(f"Probabilidad de error para umbral 1: {Pe_1:.4f}")
print(f"Probabilidad de error para umbral 2: {Pe_2:.4f}")
```

```python
plt.figure(figsize=(10,5))
plt.hist(Y[H == 0], bins=np.arange(-0.5, 3.5, 0.5), alpha=0.5, label='P(Y|H=0)', density=True)
plt.hist(Y[H == 1], bins=np.arange(-0.5, 3.5, 0.5), alpha=0.5, label='P(Y|H=1)', density=True)
plt.xticks([0, 1, 2])
plt.xlabel('Valor de Y')
plt.ylabel('Probabilidad')
plt.legend()
plt.title('Distribución de Y condicionado a H')
plt.grid()
plt.show()
```
