Considere una variable aleatoria  $X \in \{-1, +3\}$  con probabilidades a priori  $P_X(-1) = 0.4$ y  $P_X (+3) = 0.6$  . Dada una observación  y  de la variable aleatoria  Y = X+N , donde  N  es una variable aleatoria gaussiana de media  $\mu =0$  con varianza  $\sigma^2 = 0.4$ , independiente de  X , 




#### 1. Encuentre el nivel de decisión para un detector MAP.



El detector MAP decide el valor $\hat{X}(y)$ que maximiza la probabilidad a posteriori:
$$
\hat{X}(y) = \underset{x \in \{-1,+3\}}{\operatorname{argmax}} \; \Big\{ f_{Y|X}(y|x) \, P_X(x) \Big\}.
$$

La función de densidad condicional de $Y$ dado que $X=x$ es:
$$
f_{Y|X}(y|x) = \frac{1}{\sqrt{2\pi (0.4)}} \exp\Bigg(-\frac{(y-x)^2}{2 \times 0.4}\Bigg).
$$

Por lo tanto, para comparar las dos hipótesis, decido que $x=+3$ si
$$
f_{Y|X}(y|+3)\, P_X(+3) > f_{Y|X}(y|-1)\,P_X(-1),
$$
y se decide $x=-1$ en caso contrario.



Sustituyendo la función de densidad gaussiana y los valores a priori, la desigualdad se escribe como:
$$
\frac{1}{\sqrt{2\pi (0.4)}} \exp\Bigg(-\frac{(y-3)^2}{0.8}\Bigg) \times 0.6 > \frac{1}{\sqrt{2\pi (0.4)}} \exp\Bigg(-\frac{(y+1)^2}{0.8}\Bigg) \times 0.4.
$$

Al cancelar el factor común $\frac{1}{\sqrt{2\pi (0.4)}}$, se tiene:
$$
0.6 \exp\Bigg(-\frac{(y-3)^2}{0.8}\Bigg) > 0.4 \exp\Bigg(-\frac{(y+1)^2}{0.8}\Bigg).
$$



Ahora dividimos ambos lados de la desigualdad por $0.4$:
$$
1.5 \exp\Bigg(-\frac{(y-3)^2}{0.8}\Bigg) > \exp\Bigg(-\frac{(y+1)^2}{0.8}\Bigg).
$$

Tomando el logaritmo natural en ambos lados:
$$
\ln(1.5) -\frac{(y-3)^2}{0.8} > -\frac{(y+1)^2}{0.8}.
$$

Reorganizando los términos, se obtiene:
$$
\ln(1.5) > \frac{(y-3)^2 - (y+1)^2}{0.8}.
$$

Calculemos la diferencia:
$$
(y-3)^2 - (y+1)^2 = \Big(y^2 - 6y + 9\Big) - \Big(y^2 + 2y + 1\Big) = -8y + 8 = 8(1-y).
$$

Sustituyendo en la desigualdad:
$$
\ln(1.5) > \frac{8(1-y)}{0.8} = 10(1-y).
$$

Es decir,
$$
10(1-y) < \ln(1.5).
$$

Dividiendo por $10$:
$$
1-y < \frac{\ln(1.5)}{10}.
$$

Despejando $y$ se obtiene:
$$
y > 1 - \frac{\ln(1.5)}{10}.
$$

Evaluando numéricamente:
- $\ln(1.5) \approx 0.4055$,  
- $\frac{\ln(1.5)}{10} \approx 0.04055$,  
- Por lo tanto, el nivel de decisión es:
$$
y > 1 - 0.04055 \approx 0.95945.
$$

**Regla de Decisión MAP:**
- Si $y > 0.96$ (aproximadamente), se decide que $\hat{X}(y)=+3$.
- Si $y \leq 0.96$, se decide que $\hat{X}(y)=-1$.


#### 2. Calcule analíticamente la probabilidad de error   $P_e$.

La probabilidad de error total es:
$$
P_e = P(\text{error}|X=-1) \, P_X(-1) + P(\text{error}|X=+3) \, P_X(+3).
$$

Bajo la Hipótesis $X=-1$:
- Se tiene $Y\sim \mathcal{N}(-1,0.4)$.
- El error ocurre cuando se decide $+3$, es decir, cuando $y>0.96$.
- La probabilidad condicional de error es:
$$
P(\text{error}|X=-1) = P\Big(Y>0.96 \,\Big|\, X=-1\Big) = Q\!\left(\frac{0.96 - (-1)}{\sqrt{0.4}}\right) = Q\!\left(\frac{1.96}{\sqrt{0.4}}\right).
$$

Bajo la Hipótesis $X=+3$:
- Se tiene $Y\sim \mathcal{N}(3,0.4)$.
- El error ocurre cuando se decide $-1$, es decir, cuando $y\leq 0.96$.
- La probabilidad condicional de error es:
$$
P(\text{error}|X=+3) = P\Big(Y\leq 0.96 \,\Big|\, X=+3\Big) = \Phi\!\left(\frac{0.96 - 3}{\sqrt{0.4}}\right) = Q\!\left(\frac{3-0.96}{\sqrt{0.4}}\right).
$$

Por lo tanto, la probabilidad total de error es:
$$
P_e = 0.4 \; Q\!\left(\frac{1.96}{\sqrt{0.4}}\right) + 0.6 \; Q\!\left(\frac{3-0.96}{\sqrt{0.4}}\right).
$$
